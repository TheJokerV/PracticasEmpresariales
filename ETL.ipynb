<h1>CAMBIO DE DATOS</h1>

```python
base.dtypes
```

```python
base = base.replace({'FECHA_EGRESO' : '[NINGUNO]'}, np.nan)
```

```python
base['FECHA_EGRESO'] = pd.to_datetime(base['FECHA_EGRESO'])
```

```python
base['FECHA_NACIMIENTO'] = pd.to_datetime(base['FECHA_NACIMIENTO'])
```

```python
base['GENERO'] = base['GENERO'].astype('object')
```

```python
base.GENERO = base.GENERO.replace({0 : 'FEMENINO', 1 : 'MASCULINO'})
```
